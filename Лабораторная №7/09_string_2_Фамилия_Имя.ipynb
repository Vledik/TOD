1. Считайте слова из файла `litw-win.txt` и запишите их в список `words`. В заданном предложении исправьте все опечатки, заменив слова с опечатками на ближайшие (в смысле расстояния Левенштейна) к ним слова из списка `words`. Считайте, что в слове есть опечатка, если данное слово не содержится в списке `words`. 

```python
text = '''с велечайшим усилием выбравшись из потока убегающих людей Кутузов со свитой уменьшевшейся вдвое поехал на звуки выстрелов русских орудий'''
```

```python
with open('litw-win.txt', 'r', encoding='windows-1251') as file:
    words = file.read().split()

sentence ="с велечайшим усилием выбравшись из потока убегающих людей Кутузов со свитой уменьшевшейся вдвое поехал на звуки выстрелов русских орудий"

def correct_typos(sentence, words):
    corrected_sentence = []
    for word in nltk.word_tokenize(sentence):
        if word not in words:
            closest_word = min(words, key=lambda x: edit_distance(word, x))
            corrected_sentence.append(closest_word)
        else:
            corrected_sentence.append(word)
    return ' '.join(corrected_sentence)

corrected_sentence = correct_typos(sentence, words)
print(corrected_sentence)
```

1.3 Напишите функцию, которая для заданного слова `word` возвращает `k` ближайших к нему слов из списка `words` (близость слов измеряется с помощью расстояния Левенштейна)

```python
from typing import Set, List

def get_similar_words(word: str, k: int, words: Set[str]) -> List[str]:
    """Функция для возврата k подобных слов для word из коллекции words_data"""
    buf_tuple = [(edit_distance(word, item), item) for item in words]
    buf_tuple.sort(key=lambda x: x[0])
    return [x[1] for x in buf_tuple[:k]]

get_similar_words("seedless", 11, unique_words_set)
```

```python
from scipy.spatial.distance import cosine
```

```python
data
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(data["preprocessed_descriptions"])

similarity_matrix = []
for i in range(len(data)):
    row = []
    for j in range(len(data)):
        if i == j:
            row.append(1.0)
        else:
            cosine_distance = cosine(tfidf_matrix[i].toarray()[0], tfidf_matrix[j].toarray()[0])
            similarity = 1 - cosine_distance
            row.append(similarity)
    similarity_matrix.append(row)
similarity_df = pd.DataFrame(similarity_matrix, columns=data['name'], index=data['name'])
similarity_df
```
